```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
aldat = pd.read_csv('aldat.dat',header=None)
agdat = pd.read_csv('agdat.dat',header=None)

al = aldat[2].as_matrix()
ag = agdat[2].as_matrix()

Es = np.linspace(8,121,121-7)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=1,figsize = (14,7))

ax.semilogy(Es,al,label='Al')
ax.semilogy(Es,ag,label='Ag')
ax.set_title('Energy dependent attenuation',fontsize=16)
ax.legend(loc=0,fontsize=16)
ax.set_xlabel('KeV',fontsize=16)
ax.set_ylabel('X-ray attenuation',fontsize=16);
```

```python
fig,ax = plt.subplots(nrows=1,ncols=1,figsize = (14,7))

ax.plot(Es,aldat[1].as_matrix())
ax.set_title('X-ray spectrum',fontsize=16)
ax.set_xlabel('KeV',fontsize=16)
ax.set_ylabel('X-ray intensity',fontsize=16);
```
